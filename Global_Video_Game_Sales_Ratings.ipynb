```python
final_data['Year_of_Release']=final_data['Year_of_Release'].astype(int)
final_data['NA_Sales']=final_data['NA_Sales'].astype(float)
final_data['EU_Sales']=final_data['EU_Sales'].astype(float)
final_data['JP_Sales']=final_data['JP_Sales'].astype(float)
final_data['Other_Sales']=final_data['Other_Sales'].astype(float)
final_data['Global_Sales']=final_data['Global_Sales'].astype(float)
final_data['Critic_Score']=final_data['Critic_Score'].astype(int)
final_data['Critic_Count']=final_data['Critic_Count'].astype(int)
final_data['User_Count']=final_data['User_Count'].astype(int)
```

```python
final_data['User_Score']=final_data['User_Score']*10
```

```python
year=final_data["Genre"].value_counts().head(10)
print(year)
plt.figure(figsize=(8,8))
plt.pie(year,labels=final_data["Genre"].value_counts().head(10).index,autopct="%0.0f%%",shadow=True)
plt.show()
```

```python
final_data["Limited"]=~final_data.duplicated(subset=["Name"],keep=False)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
axes[-1].remove()

for index, Sales in enumerate(["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]):
    sns.barplot(data= final_data, x=Sales,y="Genre",
                estimator=sum,
                ci=None,
                ax=axes[index]).set_title("Video game sales in "+Sales.split("_Sales")[0])
   
plt.suptitle("Video game sales by region and genre", fontsize=16)
plt.tight_layout()
plt.show()
```
